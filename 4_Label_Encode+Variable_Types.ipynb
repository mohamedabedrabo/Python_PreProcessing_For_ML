# Label Endocing - 2 values label

```python
# Converting labels into numeric form on order to perform calssification
# Machine learning models can't handle : infected, not infected. it should be : 0, 1
```

### 1. Import Libraries

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

### 2. Data import and first look

```python
cancer_data = pd.read_csv('data.csv')
```

```python
# The label is 'diagnosis' column
cancer_data.head()
```

```python
# We can see that we have 2 values :
# 'B' for Benign (0) , 'M' for Malignant (1)
cancer_data.diagnosis.value_counts()
```

 ### 3. Encode

#### Perform encoding:

```python
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
print(labels)
```

#### How did M got 1 and B got 0:

```python
# Because B comes first in Alphabet.
```

#### Append to df:

```python
cancer_data['target'] = labels
```

```python
cancer_data.head()
```

#### Drop diagnosis cloumn:

```python
cancer_data.drop(columns = ['diagnosis'])
```
